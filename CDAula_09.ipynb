##LogNormal
**1.Gere um plot da densidade de probabilidade da distribuição e de sua cumulativa.**

```python
x=np.linspace(0,6,200)
shape, loc, scale = stats.lognorm.fit(x, floc=0)
lognorm_cdf = stats.lognorm.cdf(x, shape, loc=0, scale=scale) #FDA
lognorm_pdf = stats.lognorm.pdf(x, shape, scale=scale)

plt.plot(x,lognorm_pdf,'r')
```

```python
plt.plot(x,lognorm_cdf,'r')
```
