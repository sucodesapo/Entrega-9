##LogNormal
**1.Gere um plot da densidade de probabilidade da distribuição e de sua cumulativa.**

```python
import math
x=np.linspace(0,6,200)

lognorm_cdf = stats.lognorm.cdf(x, 0.5) #FDA
lognorm_pdf = stats.lognorm.pdf(x,0.5)

plt.plot(x,lognorm_pdf,'r',label="0.5")
plt.plot(x,stats.lognorm.pdf(x,0.8),'g',label="0.8")
plt.plot(x,stats.lognorm.pdf(x,1.5),'b',label="1.5")
plt.legend()
plt.show()

```

```python
plt.plot(x,lognorm_cdf,'r',label="0.5")

plt.plot(x,stats.lognorm.cdf(x,0.8),'g',label="0.8")
plt.plot(x,stats.lognorm.cdf(x, 1.5),'b',label="1.5") #FDA
plt.legend()
plt.show()



```

## t de Student
**1.Gere um plot da densidade de probabilidade da distribuição e de sua cumulativa.**

```python
x2 = np.linspace(-10, 10, 1000)

tstudent_cdf = stats.t.cdf(x2, 2, loc=0, scale=1)
tstudent_pdf = stats.t.pdf(x2, 2, loc=0, scale=1)


plt.plot(x2,tstudent_pdf,'r',label="1")
plt.plot(x2,stats.t.pdf(x2, 2, loc=0, scale=1.5),color='g',label="1.5")
plt.plot(x2,stats.t.pdf(x2, 2, loc=0, scale=1.8),color='b',label="1.8")
plt.legend()
plt.show()

```

```python
plt.plot(x2,tstudent_cdf,'r',label="1")
plt.plot(x2,stats.t.cdf(x2, 2, loc=0, scale=1.5),'g',label="1.5")
plt.plot(x2,stats.t.cdf(x2, 2, loc=0, scale=1.8),'b',label="1.8")
plt.legend()
plt.show()
```
